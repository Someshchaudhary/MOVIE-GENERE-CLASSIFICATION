```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MultiLabelBinarizer
import re
```

```python
# Using a raw string for the file path
df = pd.read_csv(r"C:\Users\Somesh\Downloads\archive\Genre Classification Dataset\train_data.txt", sep=':::', names=['title', 'genre', 'plot'], engine="python")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Preprocess the text data: lowercasing, removing punctuation, etc.
# You can use nltk or other libraries for text preprocessing, but for simplicity, we will assume basic clean data

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

# Transform the 'plot' column into TF-IDF features
X_tfidf = tfidf_vectorizer.fit_transform(df['plot'])

# Target variable
y = df['genre']

# Split data into training and test sets
# X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)

# Train a logistic regression model
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_tfidf, y)

# # Make predictions and evaluate the model
# y_pred = lr_model.predict(X_test)
# print("Accuracy with TF-IDF and Logistic Regression:", accuracy_score(y_test, y_pred))

```

```python
# Path to test data and solution
test_data_path = r"C:\Users\Somesh\Downloads\archive\Genre Classification Dataset\test_data.txt"
test_solution_path = r"C:\Users\Somesh\Downloads\archive\Genre Classification Dataset\test_data_solution.txt"

# Load test data (only movie_name and plot)
test_df = pd.read_csv(test_data_path, sep=":::", names=['movie', 'plot'], engine='python')

# Load the solution data (movie_name, genre, plot)
test_solution_df = pd.read_csv(test_solution_path, sep=":::", names=['movie', 'genre', 'plot'], engine='python')

# Ensure that the movie names match between test data and solutions
assert test_df['movie'].equals(test_solution_df['movie']), "Movie names don't match between test data and solution!"
```

```python
# Assuming TF-IDF vectorizer is already fitted (from previous steps)
X_test_tfidf = tfidf_vectorizer.transform(test_df['plot'])

# Predict genres for the test data using the trained Logistic Regression model
y_test_pred_tfidf = lr_model.predict(X_test_tfidf)

# Evaluate the accuracy using the true labels
accuracy_tfidf = accuracy_score(test_solution_df['genre'], y_test_pred_tfidf)
print("Accuracy with TF-IDF:", accuracy_tfidf)
```
